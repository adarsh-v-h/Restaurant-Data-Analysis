```python
df.isnull().sum()
# Show no null values except the Cuisines, which we will put as "unknown" for now
df['Cuisines'] = df['Cuisines'].fillna("Unknown")
```

```python
df.duplicated().sum()
```

```python
# Now we have to categorical count, for the columns
Cat_columns =  ['City', 'Country Code', 'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']
for col in Cat_columns:
    print(f"\n----{col}-----")
    print(df[col].value_counts())
```

```python
# Based on all analysis, we will drop a few 0 or almost 0 relevant columns
droppers = ["Restaurant ID",
"Restaurant Name",
"Address",
"Locality",
"Locality Verbose",
"Currency",
"Switch to order menu",
"Is delivering now",
"Rating color",
"Rating text"]
df = df.drop(columns=droppers)
df.shape
```

```python
df.head()
```

```python
# In column "Has Table booking" and "Has Online delivery", the values are No or Yes, we will convert them as Yes: 1 and No: 0
for c in ["Has Table booking", "Has Online delivery"]:
    df[c] = df[c].map({"Yes": 1, "No": 0})
```

```python
df.head()
```

```python
df['Cuisines'].head(25)
# This shows us that alot of restarunts with more than 1 Cuisines and most them repetative, we will create a new column called as primary Cuisine, where we will keep the first cuisine only for each restarunt
df["Primary Cuisine"] = df["Cuisines"].str.split(",").str[0].str.strip()
```

```python
# Now lets see where we stand
df.head()
```

```python
df["Primary Cuisine"].value_counts()
```

```python
city_freq = df['City'].value_counts() # Hold the frequence of each city
df['City_Freq'] = df['City'].map(city_freq)
df.head()
```

```python
# now have to do some chnages with the country code, so that ML doesn't think it of some other value, so have to put it in category
df['Country Code'] = df['Country Code'].astype('category').cat.codes #.cat.codes to generate integer values
```

```python
# Here a lot of cities are less than 10, so we will replace all them and give them one name as other..
count = df["Primary Cuisine"].value_counts()
top = count[count >= 10].index
df['Cuisine_Cleaned'] = df['Primary Cuisine'].apply(
    lambda x: x if x in top else "Other"
)
```
